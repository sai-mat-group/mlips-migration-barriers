```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Load CSV files
file2 = "/Users/achinthyakrishna07/Downloads/MLIP_NEB_outputs/reshma_data/structure_analysis_M3GNnet_run.csv"
file1 = "/Users/achinthyakrishna07/Downloads/MLIP_NEB_outputs/reshma_data/structure_analysis_interpola.csv"
plot_output_dir = "/Users/achinthyakrishna07/Downloads/migration_barrier/plots/structure_analysis_tight_run_eb_v_interpola.png"  # Replace with actual output directory
output_file = '/Users/achinthyakrishna07/Downloads/MLIP_NEB_outputs/reshma_data/structure_analysis_M3GNnet_run_v_interpola.csv'
# Ensure the output directory exists
os.makedirs(plot_output_dir, exist_ok=True)

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Merge the two datasets on Structure and Index
merged_df = df1.merge(df2, on=["Structure", "Index"], suffixes=("_sys1", "_sys2"))

# Compute differences (df2 - df1)
merged_df["Diff_Mean_Dist"] = merged_df["Mean Distance Difference (Å)_sys2"] - merged_df["Mean Distance Difference (Å)_sys1"]
merged_df["Diff_Max_Dist"] = merged_df["Max Distance Difference (Å)_sys2"] - merged_df["Max Distance Difference (Å)_sys1"]
merged_df["Diff_Mean_Solid_Angle"] = merged_df["Mean Solid Angle Difference (sr)_sys2"] - merged_df["Mean Solid Angle Difference (sr)_sys1"]
merged_df["Diff_Max_Solid_Angle"] = merged_df["Max Solid Angle Difference (sr)_sys2"] - merged_df["Max Solid Angle Difference (sr)_sys1"]

# Find the maximum of these differences
merged_df["Max_Diff"] = merged_df[["Diff_Mean_Dist", "Diff_Max_Dist", "Diff_Mean_Solid_Angle", "Diff_Max_Solid_Angle"]].max(axis=1)

# Assign labels
conditions = [
    (merged_df["Max_Diff"] > 0.1),
    (merged_df["Max_Diff"] < 0.01)
]
labels = ["Bad", "Good"]
merged_df["Label"] = np.select(conditions, labels, default="Comparable")

# Save the max difference details to file
max_diff_df = merged_df[["Structure", "Index", "Max_Diff", "Label"]]
max_diff_df.to_csv(output_file, index=False)

# Separate data based on labels
good_data = merged_df[merged_df["Label"] == "Good"]["Index"]
bad_data = merged_df[merged_df["Label"] == "Bad"]["Index"]
comparable_data = merged_df[merged_df["Label"] == "Comparable"]["Index"]

# Plot histograms
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Good Histogram
axes[0].hist(good_data, bins=20, color='green', alpha=0.7, edgecolor='black')
axes[0].set_title("Good")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Frequency")
fig.savefig(os.path.join(plot_output_dir, "histogram_good.png"))

# Bad Histogram
axes[1].hist(bad_data, bins=20, color='red', alpha=0.7, edgecolor='black')
axes[1].set_title("Bad")
axes[1].set_xlabel("Index")
fig.savefig(os.path.join(plot_output_dir, "histogram_bad.png"))

# Comparable Histogram
axes[2].hist(comparable_data, bins=20, color='blue', alpha=0.7, edgecolor='black')
axes[2].set_title("Comparable")
axes[2].set_xlabel("Index")
fig.savefig(os.path.join(plot_output_dir, "histogram_comparable.png"))

# Adjust layout
plt.tight_layout()
plt.show()

# Print total counts
total_counts = merged_df["Label"].value_counts()
print("Total Counts:")
print(total_counts)
```

```python
import os
import numpy as np
import csv
from pymatgen.core import Structure

# Define all structure paths
#all_structures_path = {
 #   'P2S_path2': '/Users/achinthyakrishna07/Downloads/migration_barrier/P2S/path2',
  #  'P2S_path3': '/Users/achinthyakrishna07/Downloads/migration_barrier/P2S/path3',
   #'PS2_path1': '/Users/achinthyakrishna07/Downloads/migration_barrier/PS2/path1',
   # 'PS2_path2': '/Users/achinthyakrishna07/Downloads/migration_barrier/PS2/path2',
    #'PS2_path3': '/Users/achinthyakrishna07/Downloads/migration_barrier/PS2/path3',
    #'PSi2_path1': '/Users/achinthyakrishna07/Downloads/migration_barrier/PSi2/path1',
    #'PSi2_path2': '/Users/achinthyakrishna07/Downloads/migration_barrier/PSi2/path2',
    #'PSi2_path3': '/Users/achinthyakrishna07/Downloads/migration_barrier/PSi2/path3',
#}
all_structures_path = {
    'test': '/Users/achinthyakrishna07/Downloads/migration_barrier/P2Si'
}
def compute_cartesian_difference(structure_file_1, structure_file_2):
    """Computes Cartesian coordinate difference of the 0th species between two structures."""
    structure_1 = Structure.from_file(structure_file_1)
    structure_2 = Structure.from_file(structure_file_2)
    
    coords_1 = np.array(structure_1[0].coords)
    coords_2 = np.array(structure_2[0].coords)
    
    difference = np.linalg.norm(coords_1 - coords_2)
    
    return difference

# CSV Output File
output_csv = "/Users/achinthyakrishna07/Downloads/ca_diff_test.csv"

# Open CSV file and write headers before looping through structures
with open(output_csv, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Structure", "Index", "Cartesian Difference (Å)"])

    # Loop through all structures and indices (j=1 to 8)
    for structure_name, base_path in all_structures_path.items():
        for j in range(1, 8):  # j goes from 1 to 8
            dft_structure_path = os.path.join(base_path, f"DFT_NEB_structures/{j}/CONTCAR.vasp")
            mace_structure_path = os.path.join(base_path, f"tight_tangent/neb_test_{j}.vasp")

            if os.path.exists(dft_structure_path) and os.path.exists(mace_structure_path):
                difference = compute_cartesian_difference(dft_structure_path, mace_structure_path)
                writer.writerow([structure_name, j, difference])
                print(f"Processed {structure_name}, Index {j}")
            else:
                print(f"Skipping {structure_name}, Index {j} (Missing Files)")

print(f"\n✅ Results saved to {output_csv}")
```

```python
## import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = "/Users/achinthyakrishna07/Downloads/Mace_tight_run _Ca_posn_comparison.csv"
df = pd.read_csv(file_path)

# Rename columns properly
df.columns = [
    "Structure", "Index", "Cartesian Difference (Å)", "0.1,std", "5,tangent",
    "5,eb", "2.5,std", "2.5,tangent", "2.5,eb", "0.1,tangent", "0.1,eb", "Duplicate"
]

# Drop the first row, which contained misaligned headers
df = df.iloc[1:].reset_index(drop=True)

# Convert numeric columns to float
numeric_cols = df.columns[2:-1]  # Excluding Structure and Duplicate column
df[numeric_cols] = df[numeric_cols].astype(float)

# Compute how often "Cartesian Difference (Å)" is NOT the minimum
count_not_min = {}
for col in numeric_cols[1:]:  # Exclude "Cartesian Difference (Å)"
    count_not_min[col] = (df["Cartesian Difference (Å)"] > df[col]).sum()

# Print number of cases where "Cartesian Difference (Å)" is not the minimum
print("Number of cases where 'Cartesian Difference (Å)' is NOT the minimum:")
for method, count in count_not_min.items():
    print(f"{method}: {count}")

# Identify which index had the minimum value most frequently
df["Min Method"] = df[numeric_cols].idxmin(axis=1)
index_counts = df["Index"].astype(int).value_counts().sort_index()

# Plot Histogram
plt.figure(figsize=(8, 5))
plt.bar(index_counts.index, index_counts.values, color='skyblue', edgecolor='black')

plt.xlabel("Index")
plt.ylabel("Frequency (Number of Cases with Minimum Value)")
plt.title("Frequency Distribution of Minimum Values per Index")
plt.xticks(range(1, 8))  # Ensure x-axis has indices 1 to 7
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
